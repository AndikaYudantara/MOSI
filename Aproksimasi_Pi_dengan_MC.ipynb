```
import matplotlib.pyplot as plt
import math
import numpy as np
```

```
# number of try
trial = 100
# N +diff every trial
diff = 100

arrPi = []
arrTry = []
N = [diff]
r = 1
Try = 0


for i in range(trial):
  N.append(N[i]+diff)
  z = 0
  for j in range(N[i]):
    x = np.random.rand()*r
    y = np.random.rand()*r
    if (x*x + y*y) <= r:
      z += 1
  
  pi = 4*(z/N[i])
  arrPi.append(pi)
  Try += 1
  arrTry.append(Try)
```

```
plt.plot(arrTry, arrPi)
plt.xlabel('N (x100)')
plt.ylabel('Approx Pi')
plt.show()
```
